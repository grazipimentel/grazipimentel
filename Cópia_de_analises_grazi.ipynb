```python
import pandas as pd

# Carregar os dados de cada tabela
dados1 = pd.read_csv('dados1.csv')
dados2 = pd.read_csv('dados2.csv')
dados3 = pd.read_csv('dados3.csv')
dados4 = pd.read_csv('dados4.csv')

```

```python
# Concatenando os dados
dados_concatenado = pd.concat([dados1, dados2, dados3, dados4])
dados_concatenado
```

```python
# @title miRNA Names

from matplotlib import pyplot as plt
import seaborn as sns
dados_concatenado.groupby('miRNA Names').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title miRNA Names vs Term Genes

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(dados_concatenado['miRNA Names'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(dados_concatenado, x='Term Genes', y='miRNA Names', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
# @title miRNA Names vs Target Genes (n)

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(dados_concatenado['miRNA Names'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(dados_concatenado, x='Target Genes (n)', y='miRNA Names', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
# Importar pandas e numpy
import pandas as pd
import numpy as np

# Inicializar listas para armazenar as correlações
correlation_miRNA_gene = []
correlation_miRNA_target_gene = []

```

```python
# Iterar sobre as linhas do DataFrame
for index, row in dados_concatenado.iterrows():
    # Selecionar as variáveis relevantes para a correlação
    miRNA = float(row['miRNAs (n)'])
    term_genes = float(row['Term Genes'])
    target_genes = float(row['Target Genes (n)'])

```

```python
# prompt: calcular a correlação entre miRNA e expressão genica

# Calcular a correlação entre miRNA e expressão gênica
correlation_miRNA_gene.append(np.corrcoef(miRNA, term_genes)[0, 1])
```

```python
# prompt: calcular a correlação ente miRNA e target genes

# Calcular a correlação entre miRNA e target genes
correlation_miRNA_target_gene.append(np.corrcoef(miRNA, target_genes)[0, 1])
```

```python
print(len(correlation_miRNA_gene))
print(len(correlation_miRNA_target_gene))
```

```python
# Visualizar as correlações entre miRNA e expressão gênica
print("Correlações entre miRNA e expressão gênica:")
print(correlation_miRNA_gene)

# Visualizar as correlações entre miRNA e target genes
print("\nCorrelações entre miRNA e target genes:")
print(correlation_miRNA_target_gene)
```
